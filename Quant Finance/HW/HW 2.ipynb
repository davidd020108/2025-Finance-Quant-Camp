# Modeling Correlated Stocks with Geometric Brownian Motion

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_style('darkgrid')
```

The following code is used to model stock paths as a Geometric Brownian Motion

```python
def GBM_paths(S0, sigma, t, r, mu, n_sims, n_steps):
    noise = np.random.normal(loc = 0, scale = 1, size = (n_sims, n_steps))

    dt = t/n_steps

    log_returns = (mu+r - .5*sigma**2)*dt + sigma*np.sqrt(dt)*noise

    exponent = np.cumsum(log_returns, axis = 1)

    paths = S0*np.exp(exponent)

    paths_from_start = np.insert(paths, 0, S0, axis = 1)
    
    return paths_from_start
    
```

```python
### Example

S0 = 300
sigma = .456
t = 1
r = 0
mu = 0
n_sims = 1
n_steps = 252

stock_paths = GBM_paths(S0,sigma,t,r,mu,n_sims,n_steps)


plt.figure(figsize = (10,4))

for path in stock_paths:
    plt.plot(path)

plt.title('Simulated stock path as a Geometric Brownian Motion')
plt.show()
```

### Value at Risk (VaR)

**Definition:**  
Value at Risk (VaR) at confidence level $\alpha \in (0,1)$ for a portfolio (or random loss $L$) is the threshold number $v_\alpha$ such that the probability of losses exceeding $v_\alpha$ is at most $1-\alpha$:

$$
\text{VaR}_\alpha(L) = \inf \{ \ell \in \mathbb{R} : \Pr(L \leq \ell) \geq \alpha \}.
$$

---

**Interpretation:**  
- With probability $\alpha$, the loss will **not** exceed $\text{VaR}_\alpha(L)$.  
- With probability $1-\alpha$, the loss **may exceed** this amount.  
- In practice, $\text{VaR}_\alpha$ is the $\alpha$-quantile of the loss distribution.  

---

**Example:**  
If daily portfolio losses are modeled, then the **95% VaR** is the 5th percentile of the loss distribution.  
- If $\text{VaR}_{0.95} = \$2\text{M}$, this means: *‚ÄúThere is only a 5% chance that daily losses exceed \$2M.‚Äù*  


If $X$ and $Y$ are random variables, then  

$$
\mathrm{Var}(X+Y) \;=\; \mathrm{Var}(X) + \mathrm{Var}(Y) + 2\big(E[XY]-E[X]E[Y]\big).
$$

The term $E[XY]-E[X]E[Y]$ is the **covariance** of $X$ and $Y$, denoted $\mathrm{Cov}(X,Y)$.  
The **correlation** of $X$ and $Y$ is then  

$$
\mathrm{Corr}(X,Y) = \frac{\mathrm{Cov}(X,Y)}{\mathrm{std}(X)\,\mathrm{std}(Y)}.
$$
